# Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# extraindo dados do arquivo

```python
df = pd.read_excel('DataSet_Test.xlsx')
```

# extraindo dados para a analise

```python
paises_df = df.groupby('Country')['BillNo'].count().reset_index()
customers_df = df.groupby('CustomerID')['BillNo'].count().reset_index()
```

# Gerando gr√°ficos

```python
janela_dados_fixos, graficos = plt.subplots(3, 1, figsize=(20, 10))

graficos[0].bar(paises_df['Country'], paises_df['BillNo'])
graficos[1].bar(customers_df['CustomerID'], customers_df['BillNo'])

janela_dados_fixos.tight_layout()
```
