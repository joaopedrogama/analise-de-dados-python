# Tratando dados para a an√°lise

```python
import pandas as pd
```

```python
df_vendas = pd.read_excel('../DataSet/DataSet_Test.xlsx')
```

```python
df_vendas.head(20)
```

## Tirando registro com a coluna country 'Unspecified'

```python
#Estes registro s√£o dados que n√£o se incluem na analise pois n√£o s√£o vendas de fato
df_lixo = (df_vendas['Country'] == 'Unspecified')
df_vendas_novo = df_vendas[~df_lixo]
df_vendas_novo
```

## Tirando valores nulos

```python
df_vendas_novo.dropna(inplace=True, how='any')
df_vendas_novo
```

## Tirando as horas da coluna Date

```python
df_vendas_novo['Date'] = pd.to_datetime(df_vendas_novo['Date'].dt.strftime('%d/%m/%y')) 
df_vendas_novo
```

## Exportando novo DataSet

```python
df_vendas_novo.to_excel('../DataSet/DataSetNovo.xlsx')
```
